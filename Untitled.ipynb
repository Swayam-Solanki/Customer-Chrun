```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Step 1: Diagnose the issue
def diagnose_encoding_issue(df, target_col='churned'):
    """
    Diagnose why encoding is failing
    """
    print("=== DIAGNOSIS ===")
    print(f"Dataset shape: {df.shape}")
    print(f"Target column '{target_col}' info:")
    print(f"  Data type: {df[target_col].dtype}")
    print(f"  Unique values: {df[target_col].unique()}")
    print(f"  Value counts:\n{df[target_col].value_counts(dropna=False)}")
    print(f"  Missing values: {df[target_col].isnull().sum()}")
    
    # Check for whitespace or case issues
    if df[target_col].dtype == 'object':
        print(f"  Stripped unique values: {[str(x).strip() for x in df[target_col].unique()]}")

# Step 2: Robust encoding function
def robust_encode_churned(df, target_col='churned'):
    """
    Robust encoding that handles various edge cases
    """
    df_copy = df.copy()
    
    print(f"Before encoding - {target_col} unique values: {df_copy[target_col].unique()}")
    
    # Handle string columns with potential whitespace/case issues
    if df_copy[target_col].dtype == 'object':
        # Strip whitespace and convert to lowercase for comparison
        df_copy[target_col] = df_copy[target_col].astype(str).str.strip().str.lower()
        print(f"After cleaning - {target_col} unique values: {df_copy[target_col].unique()}")
    
    # Comprehensive mapping
    mapping = {
        'yes': 1,
        'no': 0,
        '1': 1,
        '0': 0,
        1: 1,
        0: 0,
        'true': 1,
        'false': 0,
        '1.0': 1,
        '0.0': 0,
        1.0: 1,
        0.0: 0
    }
    
    # Apply mapping
    df_copy[target_col] = df_copy[target_col].map(mapping)
    
    print(f"After encoding - {target_col} unique values: {df_copy[target_col].unique()}")
    print(f"Missing values after encoding: {df_copy[target_col].isnull().sum()}")
    
    # Check what values couldn't be mapped
    if df_copy[target_col].isnull().sum() > 0:
        original_unmapped = df[df_copy[target_col].isnull()][target_col]
        print(f"Values that couldn't be mapped: {original_unmapped.unique()}")
        
        # Try to fix common issues
        for idx in original_unmapped.index:
            original_val = df.loc[idx, target_col]
            print(f"Unmapped value: '{original_val}' (type: {type(original_val)})")
    
    return df_copy

# Step 3: Fixed encoding function for entire dataset
def fixed_custom_encode_telecom_data(df):
    """
    Fixed version of encoding function
    """
    df_encoded = df.copy()
    
    # Binary columns - More robust encoding
    binary_cols = ['Dependents', 'PaperlessBilling']
    for col in binary_cols:
        if col in df_encoded.columns:
            # Clean and map
            df_encoded[col] = df_encoded[col].astype(str).str.strip().str.lower()
            df_encoded[col] = df_encoded[col].map({'yes': 1, 'no': 0})
    
    # Service columns - More robust encoding
    service_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                   'TechSupport', 'StreamingTV', 'StreamingMovies']
    for col in service_cols:
        if col in df_encoded.columns:
            df_encoded[col] = df_encoded[col].astype(str).str.strip().str.lower()
            df_encoded[col] = df_encoded[col].map({
                'yes': 1, 
                'no': 0, 
                'no internet service': -1
            })
    
    # MultipleLines - Robust encoding
    if 'MultipleLines' in df_encoded.columns:
        df_encoded['MultipleLines'] = df_encoded['MultipleLines'].astype(str).str.strip().str.lower()
        df_encoded['MultipleLines'] = df_encoded['MultipleLines'].map({
            'yes': 1,
            'no': 0,
            'no phone service': -1
        })
    
    # InternetService - One-hot encoding
    if 'InternetService' in df_encoded.columns:
        df_encoded = pd.get_dummies(df_encoded, columns=['InternetService'], prefix='Internet')
    
    # Contract - Ordinal encoding
    if 'Contract' in df_encoded.columns:
        df_encoded['Contract'] = df_encoded['Contract'].astype(str).str.strip()
        contract_map = {'Month-to-month': 0, 'One year': 1, 'Two year': 2}
        df_encoded['Contract'] = df_encoded['Contract'].map(contract_map)
    
    # PaymentMethod - One-hot encoding
    if 'PaymentMethod' in df_encoded.columns:
        df_encoded = pd.get_dummies(df_encoded, columns=['PaymentMethod'], prefix='Payment')
    
    # Target variable - Use robust encoding
    if 'churned' in df_encoded.columns:
        df_encoded = robust_encode_churned(df_encoded, 'churned')
    
    return df_encoded

# Step 4: Apply the fix
print("=== APPLYING THE FIX ===")

# First diagnose
diagnose_encoding_issue(train, 'churned')

# Apply fixed encoding
train_encoded_fixed = fixed_custom_encode_telecom_data(train)

print(f"\n=== RESULTS ===")
print(f"Original train shape: {train.shape}")
print(f"Encoded train shape: {train_encoded_fixed.shape}")
print(f"Missing values in churned: {train_encoded_fixed['churned'].isnull().sum()}")
print(f"Churned value counts:\n{train_encoded_fixed['churned'].value_counts(dropna=False)}")

# Clean and prepare for modeling
train_clean = train_encoded_fixed.dropna(subset=['churned'])
print(f"Clean train shape: {train_clean.shape}")

if train_clean.shape[0] > 0:
    # Separate features and target
    X_train_full = train_clean.drop('churned', axis=1)
    y_train_full = train_clean['churned']
    
    print(f"X_train_full shape: {X_train_full.shape}")
    print(f"y_train_full shape: {y_train_full.shape}")
    
    # Now try train_test_split
    if len(y_train_full) > 0:
        X_train, X_val, y_train, y_val = train_test_split(
            X_train_full, y_train_full, 
            test_size=0.2, 
            random_state=42, 
            stratify=y_train_full
        )
        
        print(f"\n✅ SUCCESS!")
        print(f"X_train: {X_train.shape}")
        print(f"X_val: {X_val.shape}")
        print(f"y_train: {y_train.shape}")
        print(f"y_val: {y_val.shape}")
        
        print(f"\nTarget distribution:")
        print(f"y_train: {y_train.value_counts().to_dict()}")
        print(f"y_val: {y_val.value_counts().to_dict()}")
    else:
        print("❌ Still no data after cleaning")
else:
    print("❌ No data remaining after cleaning - need to investigate further")
```
